```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.metrics import r2_score
```

```python
# Features: [Kills, Assists, Deaths]

X = np.array([
    [10, 2, 5],
    [15, 4, 6],
    [20, 5, 7],
    [25, 7, 8],
    [30, 9, 10],
    [35, 10, 12],
    [40, 12, 13],
    [45, 14, 15]
])


# Target: Player Rating (out of 100)

y = np.array([55, 60, 65, 70, 75, 80, 85, 90])
```

```python
#Step 3: Train SVR Model

model = SVR(kernel='rbf', C=100, epsilon=2)
model.fit(X, y)
```

```python
#Step 4: Make Prediction

new_player = np.array([[28, 8, 9]])  # new player stats
pred = model.predict(new_player)
print("Predicted Player Rating:", round(pred[0], 2))
```

```python
#Step 5: Evaluate Model

y_pred = model.predict(X)
r2 = r2_score(y, y_pred)
print("R² Score:", round(r2, 3))
```

```python
#Step 6: Visualize Results (Optional)

plt.scatter(X[:, 0], y, color='blue', label='Actual Ratings')
plt.scatter(X[:, 0], y_pred, color='red', label='Predicted Ratings')
plt.xlabel("Kills per Match")
plt.ylabel("Player Rating")
plt.title("SVR - Player Rating Prediction (Gaming Stats)")
plt.legend()
plt.show()
```
